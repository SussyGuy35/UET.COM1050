# Quản lý điểm số

```python
grades = list(map(float, input("Nhap diem cac mon (cach nhau boi dau cach): ").split()))
avrage = sum(grades) / len(grades)
hocluc = ""
if avrage >= 9:
    hocluc = "Xuat sac"
elif avrage >= 8.5:
    hocluc = "Gioi"
elif avrage >= 7:
    hocluc = "Kha"
elif avrage >= 5.5:
    hocluc = "Trung binh"
else:
    hocluc = "Yeu"

for grade in grades:
    if grade < 4:
        hocluc = "Kem"
        break

print(f"Diem trung binh: {avrage:.2f}")
print(f"Hoc luc: {hocluc}")
```

# Tính BMI

```python
height = float(input("Nhap chieu cao (m): "))
weight = float(input("Nhap can nang (kg): "))
bmi = weight / (height ** 2)
bmi_category = ""
if bmi < 18.5:
    bmi_category = "Thieu can"
elif bmi < 25:
    bmi_category = "Binh thuong"
elif bmi < 30:
    bmi_category = "Thua can"
else:
    bmi_category = "Beo phi"

print(f"BMI cua ban: {bmi:.2f}")
print(f"Phan loai: {bmi_category}")
```

# Thu ngân hiệu sách

```python
n = int(input("Nhap so luong mat hang: "))
hang = []
for i in range(n):
    ten = input(f"Nhap ten mat hang {i+1}: ")
    soluong = int(input(f"Nhap so luong mat hang {i+1}: "))
    gia = float(input(f"Nhap gia mat hang {i+1}: "))
    hang.append((ten, gia, soluong))

thanhtien = [gia * soluong for ten, gia, soluong in hang]
tongtien = sum(thanhtien)
if tongtien >= 100:
    giamgia = 0.1 * tongtien

elif tongtien >= 50:
    giamgia = 0.05 * tongtien
else:
    giamgia = 0

tongtien = tongtien - giamgia

print(
"""=== HOA DON HIEU SACH ===
STT | Ten mat hang | SL | Don gia | Thanh tien
----------------------------------------
"""
+
"\n".join(f"{i+1:>3} | {ten:<12} | {soluong:<2} | {gia:<7.2f} | {tt}" for i, (ten, gia, soluong), tt in zip(range(n), hang, thanhtien))
+
f"""
----------------------------------------
Tong tien: {tongtien + giamgia}
Giam gia: {giamgia}
Thanh toan: {tongtien}
========================="""
)
```
